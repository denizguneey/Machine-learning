```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score

#data upload
data=pd.read_csv('../datasources/salaries.csv')
#print(data)

#data frame slicing
x=data.iloc[:,1:2]
y=data.iloc[:,2:]

#Numpy array conversion
X=x.values
Y=y.values

#linear regression
from sklearn.linear_model import LinearRegression

linearRegression=LinearRegression()
linearRegression.fit(X,Y)

# plt.scatter(X,Y,color='red')
# plt.plot(x,linearRegression.predict(X),color='blue')
# plt.show()

#polynomial regression(non-linear model)
##y=ax^0+x^1+x^2..
from sklearn.preprocessing import PolynomialFeatures
#For second degree polynomial
polynomialregression=PolynomialFeatures(degree=2)
x_polynomial=polynomialregression.fit_transform(X)
#print(x_polynomial)
##y=ax^0+bx^1+cx^2..
linearRegression2=LinearRegression()
linearRegression2.fit(x_polynomial,y)

# plt.scatter(X,Y,color='red')
# plt.plot(X,linearRegression2.predict(polynomialregression.fit_transform(X)),color='blue')
# plt.show()

#For 4th degree polynomial
polynomialregression=PolynomialFeatures(degree=4)
x_polynomial=polynomialregression.fit_transform(X)
#print(x_polynomial)
linearRegression2=LinearRegression()
linearRegression2.fit(x_polynomial,y)


# plt.scatter(X,Y,color='red')
# plt.plot(X,linearRegression2.predict(polynomialregression.fit_transform(X)),color='blue')
#plt.show()

#print(linearRegression.predict([[11]]))
#print(linearRegression.predict([[6.6]]))

#print(linearRegression2.predict(polynomialregression.fit_transform([[11]])))
#print(linearRegression2.predict(polynomialregression.fit_transform([[6.6]])))

from sklearn.preprocessing import StandardScaler

scaler1=StandardScaler()
x_scaler=scaler1.fit_transform(X)
scaler2=StandardScaler()
y_scaler=np.ravel(scaler2.fit_transform(Y.reshape(-1,1)))

from sklearn.svm import SVR

svr_regression=SVR(kernel='rbf')
svr_regression.fit(x_scaler,y_scaler)

# plt.scatter(x_scaler,y_scaler,color='red')
# plt.plot(x_scaler,svr_regression.predict(x_scaler),color='blue')

#Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
decisionTree=DecisionTreeRegressor(random_state=0)
decisionTree.fit(X,Y)

# plt.scatter(X,Y,color='red')
# plt.plot(X,decisionTree.predict(X),color='blue')
# plt.show()
#print(decisionTree.predict([[11]]))
#print(decisionTree.predict([[6.6]]))

#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
randomForest=RandomForestRegressor(n_estimators=10,random_state=0)
randomForest.fit(X,Y.ravel())

#print(randomForest.predict([[6.6]]))
# plt.scatter(X,Y,color='red')
# plt.plot(X,randomForest.predict(X),color='blue')

#R^2 calculate
print('----------------------------------')
print('Random Forest Regression R2 value')
print(r2_score(Y,randomForest.predict(X)))

print('Decision Tree R2 value')
print(r2_score(Y,decisionTree.predict(X)))

print('Support Vector Regression R2 value')
print(r2_score(y_scaler,svr_regression.predict(x_scaler)))

print('Linear Regression R2 value')
print(r2_score(Y,linearRegression.predict(X)))

print('Polynomial Regression R2 value')
print(r2_score(Y,linearRegression2.predict(polynomialregression.fit_transform(X))))

```
