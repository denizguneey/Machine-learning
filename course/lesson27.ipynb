```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('../datasources/Ads_CTR_Optimisation.csv')
#print(data)
##
import random

N=10000
d=10
sum=0
selects=[]
for n in range(0,N):
    ad=random.randrange(d)
    selects.append(ad)
    award=data.values[n,ad]
    sum=sum+award
    
plt.hist(selects)
plt.show()
```
